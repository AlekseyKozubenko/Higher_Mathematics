Нарисуйте график функции y(x) = k∙cos(x – a) + b :   для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 301)
k = [3, 5, 10]
a = [ -2, 5, 3]
b = [-7, 0, 7]
plt.grid()
for i in range(len(k)):
    y = k[i]*np.cos(x - a[i]) + b[i]
    plt.plot(x, y)
```

# 4) Решить систему уравнений
exp(x) + x∙(1 – y) = 1  
y = x2 – 1

```python
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*(1 - y) - 1)

x1, y1 = fsolve(f, (1, 1))
x2, y2 = fsolve(f, (-1, -1))

print(x1, y1)
print(x2, y2)
```
