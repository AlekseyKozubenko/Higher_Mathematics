```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0, 40, 102)

k1=1
k2=5

y1=np.cos(k1*x)
y2=np.cos(k2*x)

fig, ax = plt.subplots()

label1 = 'cos(x)'
label2 = 'cos(' + str(k2) + '*x)'

ax.plot(x, y1, label = label1)
ax.plot(x, y2, label = label2)

ax.legend()

plt.show()
```
