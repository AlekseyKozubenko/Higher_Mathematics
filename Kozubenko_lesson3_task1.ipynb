## 1) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
A = np.random.randint(-5, 5, size=10)
A
```

```python
length2 = 0
for i in A:
    length2 += i**2
length = np.sqrt(length2)
print(f'Длина вектора А:{A} равна {length}')
```

## 3) Напишите код на Python, реализующий построение графиков: 
## окружности,	эллипса, гиперболы.

```python
# ОКРУЖНОСТЬ
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 6, 6
```

```python
x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    l = i/100
    x.append(l)
    y1.append(np.sqrt(r**2-l**2))
    y2.append(-np.sqrt(r**2-l**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# эллипс
```

```python
x = []
y1 = []
y2 = []
a = 5
b = 3
for i in range(-300, 301):
    l = i/10
    x.append(l)
    y1.append(b*np.sqrt(1-l**2/a**2))
    y2.append(-b*np.sqrt(1-l**2/a**2))

plt.axis([-7, 7, -7, 7])
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# гипербола
```

```python
x = []
y1 = []
y2 = []
a = 2
b = 3
for i in range(-300, 301):
    l = i/10
    x.append(l)
    y1.append(b*np.sqrt(l**2/a**2 - 1))
    y2.append(-b*np.sqrt(l**2/a**2 - 1))

plt.axis([-7, 7, -7, 7])
plt.plot(x, y1)
plt.plot(x, y2)
```
