Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а: 
sin(а*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).
Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.

```python
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

def f(x):
    return (np.pi / x)

a = np.linspace(0.01, 0.02, 201)
x = f(a)
fig, ax = plt.subplots()
ax1 = ax.twinx()
ax.plot(a,a,color='r')
ax1.plot(a,x,color='b')
plt.plot([0.009, 0.0205],[0,0], color ='red', linewidth=0.75, linestyle ="--") 
ax.set_ylabel('Для значений "a"')
ax1.set_ylabel('Для значений "x"')
plt.grid(True)
plt.show()
```
